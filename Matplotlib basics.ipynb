```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x= np.linspace(0,5,11)
y=x**2
x
```

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Plot')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y label')
axes.set_title('Figue')
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title('INSERT PLOT')
axes1.set_title('SMALL PLOT')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.2,0.5,0.5])
axes1.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

axes[0].set_title('1')
axes[1].set_title('0')
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi=400)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x Squared')
ax.plot(x,x**3,label='X Cubed')
ax.legend(loc=(0.1,0.8))
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=2,ls='--')

ax.set_xlim([0,2])
ax.set_ylim([0,2])
```
